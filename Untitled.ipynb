```python
# csv_list = [file for file in glob.glob("*.csv")]
csv_list = ['ranking-result.csv']
for file in csv_list:
    with open(file,'r') as dest_f:
        data_iter = csv.reader(dest_f, 
                               delimiter = ',', 
                               quotechar = '"')
        data = [data for data in data_iter]
    data_array = numpy.asarray(data, dtype=numpy.float32)
    print(file)
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    sns.regplot(x=data_array[:, 0], y=data_array[:, 1], ax=ax1);
    sns.regplot(x=data_array[:, 0], y=data_array[:, 2], ax=ax2);
    plt.show()
```

```python
# csv_list = [file for file in glob.glob("*.csv")]
csv_list = ['error-change-result.csv']
for file in csv_list:
    with open(file,'r') as dest_f:
        data_iter = csv.reader(dest_f, 
                               delimiter = ',', 
                               quotechar = '"')
        data = [data for data in data_iter]
        data_array = numpy.asarray(data, dtype=numpy.float32)
        print(file)
        f, ax = plt.subplots(1, 3, figsize=(16, 6))
        ax = ax.flatten()
        for i in range(3):
            sns.tsplot(data_array[:, i], ax=ax[i]);
        plt.show()
```
